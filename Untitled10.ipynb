```python
x = [1,2,3,4,5]
y = [1.2, 1.8, 2.6, 3.2, 3.8]

n = len(x)

x_mean = sum(x) / n
y_mean = sum(y) / n

xy_sum = sum(x[i] * y[i] for i in range(n))
x_squared_sum = sum(x[i] ** 2 for i in range(n))

slope = (xy_sum - n * x_mean * y_mean) / (x_squared_sum - n * x_mean ** 2)
intercept = y_mean - slope * x_mean

x_7 = 7
y_pred_7 = slope * x_7 + intercept

x_9 = 9
y_pred_9 = slope * x_9 + intercept

print(f"Predicted value of y for x = {x_7}: {y_pred_7}")

print(f"Predicted value of y for x = {x_9}: {y_pred_9}")

# Generate points for the regression line
x_reg = [min(x), max(x)]
y_reg = [slope * xi + intercept for xi in x_reg]

# Plot the data points and the regression line
plt.scatter(x, y, label='Data Points')
plt.plot(x_reg, y_reg, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
